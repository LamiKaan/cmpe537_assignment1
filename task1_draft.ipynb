```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get directory of image files relative to this file's directory
# image_dir = os.path.join(os.path.dirname(__file__), 'data', 'images', 'task_1')
image_dir = os.path.join(os.getcwd(), 'data', 'images', 'task_1')
# Get image path based on image id
image_id = 1
image_path = os.path.join(image_dir, 'image0{}.jpeg'.format(image_id))
```

```python
# Open the image using Pillow
image = Image.open(image_path)
plt.imshow(image)

# Convert the image to a NumPy array (of shape H x W x 3)
image_matrix = np.array(image)

print(image_matrix[0, 0])

# Show the shape of the matrix
print(f"Image matrix shape: {image_matrix.shape}")

plt.show()
```
